```python
import pandas as pd
from pre_process import pre_process
from sklearn.metrics.pairwise import cosine_similarity
import pickle

df0_1 = pd.read_csv("df0_1.csv")
df0_2 = pd.read_csv("df0_2.csv")
df_score = pd.read_csv("df_score.csv")
```

```python
with open('tfidf_matrix.pkl', 'rb') as f:
    loaded_tfidf_matrix = pickle.load(f)

with open('tfidf_vectorizer.pkl', 'rb') as f:
    loaded_tfidf_vectorizer = pickle.load(f)

with open('algorithm_model.pkl', 'rb') as f:
    loaded_algorithm_model = pickle.load(f)
```

```python
test = ['Machine Learning']
test_df = pd.DataFrame({'products': test})
test_df['products'] = test_df['products'].apply(pre_process)
test_tfidf = loaded_tfidf_vectorizer.transform(test_df.products)
cosine_similarities = cosine_similarity(test_tfidf, loaded_tfidf_matrix)
df_show = pd.DataFrame(cosine_similarities)
df_show = df_show.transpose().reset_index(drop=True)
df_show = df_show.rename(columns={0: 'Similar'})
df_show = df_show.nlargest(5, 'Similar')
df_show = df_show.reset_index()
df_show = df_show.rename(columns={'index': 'CourseID'})
df_show = pd.merge(df_show, df0_1, on='CourseID', how='left')
df_show = df_show[['CourseName', 'Unit', 'Level', 'Description', 'ReviewNumber', 'AvgStar']]
df_show
```

# Collaborative Filtering

```python
userId = 72043
df_score['EstimateScore'] = df_score['CourseId'].apply(lambda x: loaded_algorithm_model.predict(userId, x).est)
df_score = df_score.sort_values(by=['EstimateScore'], ascending=False)
df_score = df_score.drop_duplicates()
df_score = pd.merge(df_score, df0_2, on='CourseId', how='left')
df_score = pd.merge(df_score, df0_1, on='CourseName', how='left')
df_score = df_score[['CourseName', 'Unit', 'Level', 'Description', 'ReviewNumber', 'AvgStar']]
df_score = df_score.head(5)
df_score
```
